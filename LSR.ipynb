```python
import numpy as np

from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd
import datetime as dt
```

```python
engine = create_engine("sqlite:///ufo.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)

# Save reference to the table
Ufos = Base.classes.ufo_sightings
```

```python
# Open session
session = Session(engine)
```

```python
session.query(Ufos.state, Ufos.date_time).all()
```

```python
state_query = session.query(Ufos.state, Ufos.date_time).all()

state_dates_frequency = {}

for result in state_query:
    state = result[0]
    dates = result[1]
    
    if state in state_dates_frequency:
        if dates in state_dates_frequency[state]:
            state_dates_frequency[state][dates] += 1
        else:
            state_dates_frequency[state][dates] = 1
    else:
        state_dates_frequency[state] = {}
    
    
#     if state_dates_frequency[dates] < '2014-01-01':
#         years.append
        
# Ufos.date_time >= '2014-01-01' and < '2015-01-01'
# Ufos.date_time >= '2015-01-01' and < '2016-01-01'
# Ufos.date_time >= '2016-01-01' and < '2017-01-01'
# Ufos.date_time >= '2017-01-01' and < '2018-01-01'
# Ufos.date_time >= '2018-01-01' and < '2019-01-01'
# Ufos.date_time >= '2019-01-01' and < '2020-01-01'
# Ufos.date_time >= '2020-01-01' and < '2021-01-01'
# Ufos.date_time >= '2021-01-01' and < '2022-01-01'
# Ufos.date_time >= '2022-01-01' and < '2023-01-01'
    
state_dates_frequency        
```
