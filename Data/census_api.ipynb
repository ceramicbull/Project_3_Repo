```python
#Dependencies
import json
import requests
import pandas as pd
#import state to postal dictionary
from us_state_abbrev import us_state_to_abbrev as abbrev
```

```python
#set url
url="https://api.census.gov/data/2021/pep/population?get=NAME,POP_2021&for=state:*"
```

```python
#make request, get response
response=requests.get(url)
data=response.json()
print(json.dumps(data, indent=4, sort_keys=True))
```

```python
#loop through data elements, replace long form state names with Postal abbreviations:
for i in data:
    if i[0] != "NAME":
        i[0]=abbrev[i[0]]
print(json.dumps(data, indent=4, sort_keys=True))
```

```python
#convert list of lists into DataFrame
df=pd.DataFrame(data, columns=data[0])
#drop duplicate header row and extraneous census state info
df.drop(index=0,columns='state',inplace=True)
df.head()
```

```python
#export CSV
df.to_csv(f"output_csv/STATE_POP.csv", index=False)
```
